```python
from datetime import datetime

#Credenciales Redshift
host = 'data-engineer-cluster.cyhh5bfevlmn.us-east-1.redshift.amazonaws.com'
port = 5439
user = 'franciscorossi48_coderhouse'
password = 'jhu3Yv805G'
database = 'data-engineer-database'

#Conexin
conn = psycopg2.connect(
    host=host,
    port=port,
    user=user,
    password=password,
    database=database
)

#Cursor
cursor = conn.cursor()

api_key = 'QSZ491MBGEV1MZG8'
symbol = 'MELI'

#Endpoint
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={symbol}&interval=1min&apikey={api_key}'

#Solicitud
response = requests.get(url)
data_json = response.json()

#Extraccin
time_series_data = data_json.get('Time Series (1min)', {})
insert_data = []

for timestamp, values in time_series_data.items():
    date_time_obj = datetime.strptime(timestamp, '%Y-%m-%d %H:%M:%S')
    insert_data.append((date_time_obj, float(values['1. open']), float(values['2. high']),
                        float(values['3. low']), float(values['4. close']), int(values['5. volume'])))

#Query
insert_query = """
INSERT INTO franciscorossi48_coderhouse.ejemplo_finanzas (fecha, apertura, alto, bajo, cierre, volumen)
VALUES (%s, %s, %s, %s, %s, %s);
"""

cursor.executemany(insert_query, insert_data)

conn.commit()
cursor.close()
conn.close()

print("Datos insertados")
```
